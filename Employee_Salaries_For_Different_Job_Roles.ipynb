```python
# Histogram
plt.figure(figsize=(5,4))
sns.histplot(df['salary'], kde=True, color='lightcoral')
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(30, 15))
sns.boxplot(x='job_title', y='salary',hue='experience_level', data=df)
plt.title('Salary Trends Across Job Titles and Experience Levels')
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.xticks(rotation=45)
plt.legend(title='Experience Level')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(5, 4))
sns.boxplot(x='employment_type', y='salary', data=df)
plt.title('Comparison of Salary Levels between Full-time and Part-time Employment')
plt.xlabel('Employment Type')
plt.ylabel('Salary')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(5, 4))
sns.scatterplot(x='remote_ratio', y='salary', data=df, color='green')
plt.title('Impact of Remote Work on Compensation')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Salary')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(5,4))
sns.barplot(x='company_size',y='salary',data=df)
plt.title('Correlation between Company Size and Employee Salaries')
plt.xlabel('Company Size')
plt.ylabel('Average Salary')
plt.show()
```

```python
# pair plot
sns.pairplot(df[['salary', 'remote_ratio', 'company_size','experience_level']], diag_kind='kde')
plt.suptitle('Pairwise Relationship between Numerical Variables', y=1.02)
plt.show()
```
