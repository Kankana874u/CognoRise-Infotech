```python
#If the number of entries is N, you have upto 50% missing values ---we can replace those rows with mean, median, mode
df['Item_Weight'] = df['Item_Weight'].fillna(df['Item_Weight'].median())


#Fill missing values in'Outlet_Size' coloumn with mode
df['Outlet_Size'] = df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0])

```

```python
# check for missing values again after filling
print(df.isnull().sum())
```

```python
# Histogram
plt.figure(figsize=(6,4))
sns.histplot(df['Item_MRP'], kde=True)
plt.title('Distribution of Item MRP')
plt.xlabel('Item MRP')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Outlet_Type', y='Item_MRP', data=df)
plt.title('MRP Distribution by Outlet Type')
plt.xlabel('Outlet Type')
plt.ylabel('Item MRP')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(8,6))
sns.countplot(x='Item_Type', data=df)
plt.title('Count of Items by Type')
plt.xlabel('Item Type')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Item_Weight', y='Item_MRP', data=df)
plt.title('Item Weight vs Item MRP')
plt.xlabel('Item Weight')
plt.ylabel('Item MRP')
plt.show()
```

```python
# pie plot
plt.figure(figsize=(6, 6))
plt.pie(df['Item_Fat_Content'].value_counts(), autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Proportion of Item Fat Content')
plt.ylabel('')


plt.show()
```

```python
# line plot
plt.figure(figsize=(7, 4))
sns.lineplot(x='Outlet_Establishment_Year', y='Item_MRP', data=df)
plt.title('Item MRP over Establishment Year')
plt.xlabel('Outlet_Establishment_Year')
plt.ylabel('Item MRP')
plt.show()
```

```python
# violin plot
plt.figure(figsize=(6, 4))
sns.violinplot(x='Outlet_Size', y='Item_MRP', data=df)
plt.title(' MRP Distribution by Outlet Size')
plt.xlabel('Outlet Size')
plt.ylabel('Item MRP')
plt.show()
```

```python
# pair plot
sns.pairplot(df[['Item_Weight', 'Item_Visibility', 'Item_MRP']], diag_kind='kde')
plt.suptitle('Pairwise Relationship between Numerical Variables', y=1.02)
plt.show()
```
